**Visualization with RDKit**

```python
from rdkit.Chem.Draw import MolsToGridImage
import rdkit.Chem as Chem

smiles=["CCCN1c2cc3nsnc3cc2N(CC)C1OC" ,"COC1N(C)c2cc3nsnc3c(OCC(C)C)c2N1C" , "CCCN1c2cc3nsnc3c(C)c2N(C)C1OC"]
rdkit_mols = [Chem.MolFromSmiles(s) for s in smiles]
MolsToGridImage(rdkit_mols,molsPerRow=4,useSVG=True)
```

**Visualization with datamol**

```python
import datamol as dm
legend_=["mol1","mol2","mol3"]
dm.to_image(rdkit_mols,outfile="mols.png",n_cols=3,use_svg=False,mol_size=(600,400),legends=legend_,legend_fontsize=20)
```
