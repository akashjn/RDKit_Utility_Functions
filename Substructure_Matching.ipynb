***Import Chem from RDKit Python Libraries***

```python
import rdkit.Chem as Chem
```

*** The function substruct_in_smi() will retrun True is the substruture pattern is found in the molecule.***

```python
def substruct_in_smi(mol_smi_str,pat_smi_str):
    """
    This function will return True if the substructure (pattern) is present in the SMILES string.
    Input: smi_str = A SMILES string;   sub_str = Pattern 
    Output: True (substruture found in the smi); False (substruture not found in the smi)
    """   
    mol = Chem.MolFromSmiles(mol_smi_str)
    pat = Chem.MolFromSmiles(pat_smi_str) 

    return mol.HasSubstructMatch(pat)
```

***Example 1.***

```python
mol_smiles_str="COCCOC1N(C)c2cc3nsnc3cc2N1C"
mol = Chem.MolFromSmiles(mol_smiles_str)
mol
```

```python
pattern_smiles_str1="COCN"
pat1 = Chem.MolFromSmiles(pattern_smiles_str1)
pat1
```

```python
print(substruct_in_smi(mol_smi_str=mol_smiles_str,pat_smi_str=pattern_smiles_str1))
```
